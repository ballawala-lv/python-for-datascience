```python
import plotly.plotly as py
from plotly.offline import download_plotlyjs, init_notebook_mode, plot,iplot
init_notebook_mode(connected=True)
```

```python
data = dict(type='choropleth', locations=['AZ', 'CA', 'NY'], locationmode='USA-states', colorscale='Jet', text=['text1', 'text2','text3'], z=[1.0,2.0,3.0],colorbar={'title':'Colorbar Title Goes Here'})
```

```python
data
```

```python
layout = dict(geo={'scope':'usa'})
import plotly.graph_objs as go
layout
```

```python
choromap = go.Figure(data=[data], layout=layout) 
iplot(choromap)
```

```python
import pandas as pd
df=pd.read_csv('2011_US_AGRI_Exports')
```

```python
df.head()
```

```python
data = dict(type='choropleth', colorscale='YlOrRd', locations=df['code'], locationmode='USA-states', z=df['total exports'], text =df['text'], colorbar={'title':'Millions USD'}, marker=dict(line=dict(color='rgb(255,255,255)', width=1)))
```

```python
layout=dict(title='2011 US Agriculture exports by state', geo=dict(scope='usa', showlakes=True, lakecolor='rgb(85,173,240)'))
```

```python
layout
```

```python
choromap2 = go.Figure(data=[data], layout=layout)
iplot(choromap2)
```

```python
df2 = pd.read_csv('2014_World_GDP')
```

```python
df2.head()
```

```python
data2 = dict(type='choropleth', locations=df2['CODE'], z=df2['GDP (BILLIONS)'], text=df2['COUNTRY'], colorbar={'title':'GDP in Billion USD'})
```

```python
layout2=dict(title='2014 Global GDP', geo=dict(showframe=False, projection={'type':'natural earth'}))
```

```python
chromap3 = go.Figure(data=[data2], layout=layout2)
iplot(chromap3)
```
